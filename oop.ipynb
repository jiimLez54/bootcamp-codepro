```python
class Animales:
    especie = "felino"   # atributo de clase

    def __init__(self, nombre, color):
        self.nombre = nombre   # atributo de instancia
        self.color = color     # atributo de instancia


# Crear objetos
animal1 = Animales("cookie", "marron")
animal2 = Animales("brow", "naranjado")

# Imprimir frase
print(f"Me llamo {animal1.nombre} y soy un {animal1.especie} de color {animal1.color}")
print(f"Me llamo {animal2.nombre} y soy un {animal2.especie} de color {animal2.color}")
```

```python
class Animales:
    especie = "ave"   # atributo de clase

    def __init__(self, nombre, color):
        self.nombre = nombre
        self.color = color

    # método hablar
    def hablar(self):
        print(f"Cuando hablo, digo larolaro capitan garciyo")

    # método moverse
    def moverse(self):
        print(f"Cuando me muevo, vuelo")


# crear objetos
animal1 = Animales("Loro", "verde")
animal2 = Animales("Canario", "amarillo")

# imprimir información
print(f"Me llamo {animal1.nombre} y soy un {animal1.especie} de color {animal1.color}")
animal1.hablar()
animal1.moverse()

print()

print(f"Me llamo {animal2.nombre} y soy un {animal2.especie} de color {animal2.color}")
animal2.hablar()
animal2.moverse()
```

```python
class Madre:
    def __init__(self, nombre, atributos):
        self.nombre = nombre
        self.atributos =atributos


class Hija(Madre):
    pass


class Hijo(Madre):
    pass


# Crear objetos de las clases hijas
persona1 = Hija("ara","pelo castaño")
persona2 = Hijo("Mati","ojos claros")

# Mostrar atributos
print("Nombre:", persona1.nombre)
print("atributos:", persona1.atributos)

print("Nombre:", persona2.nombre)
print("atributos:", persona2.atributos)
```

```python
class Heroe:
    def __init__(self, nombre, poder):
        self.nombre = nombre
        self.poder = poder

    def presentarse(self):
        print(f"Yo me llamo {self.nombre} y tengo {self.poder}")


class Villano:
    def __init__(self, nombre, poder):
        self.nombre = nombre
        self.poder = poder

    def presentarse(self):
        print(f"Yo me llamo {self.nombre} y puedo {self.poder}")


# Crear personajes
heroe1 = Heroe("Pedro", "telepatia")
villano1 = Villano("santiago", "romper objeto")

# Presentarse
heroe1.presentarse()
villano1.presentarse()
```
